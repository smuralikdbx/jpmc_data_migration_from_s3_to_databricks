```python
dbutils.widgets.text("jobId", "", "Job Id for UCD Sync  using S3 inventory service")

dbutils.widgets.dropdown("debug", "True", ["True", "False"], "Debug")

dbutils.widgets.text("s3_batch_copy_id", "Test1", "s3_batch_copy_id from the metadata mapping table")

job_id = dbutils.widgets.get("jobId")

debug = dbutils.widgets.get("debug")

dbutils.widgets.text("catalog", "users")
dbutils.widgets.text("schema", "dom_rodrigues")
dbutils.widgets.text("s3_batch_copy_id", "Test1", "Batch Copy Id to compare")

s3_batch_copy_id = dbutils.widgets.get("s3_batch_copy_id")
catalog = dbutils.widgets.get("catalog")
schema = dbutils.widgets.get("schema")


runJob=True
```

```python
if runJob:
    databricks_instance = dbutils.notebook.entry_point.getDbutils().notebook().getContext().apiUrl().getOrElse(None)

    access_token = dbutils.notebook.entry_point.getDbutils().notebook().getContext().apiToken().get()

    headers = {
        "Authorization": f"Bearer {access_token}",
        "Content-Type": "application/json"
    }

    s3_sync_task_parameters= {
        "s3_batch_copy_id": s3_batch_copy_id,
        "catalog": catalog,
        "schema": schema
    }
    
    print(f"Running sync for {s3_batch_copy_id}")
    payload = {
        "job_id": job_id,
        "notebook_params": s3_sync_task_parameters,
        "idempotency_token": str(uuid.uuid4())
    }

    response = requests.post(
        f"{databricks_instance}/api/2.0/jobs/run-now",
        headers=headers,
        json=payload
    ).json()

    print(response)
```
