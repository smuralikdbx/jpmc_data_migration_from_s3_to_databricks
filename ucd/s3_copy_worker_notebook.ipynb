```python
import json
import time
```

```python
dbutils.widgets.text("src_path", "dbfs:/tmp/file.csv", "Src File to be copied")
dbutils.widgets.text("dest_path", "dbfs:/tmp/file2.csv", "Dest File to be copied")
dbutils.widgets.text("file_type", "file", "file or dir")
dbutils.widgets.text("file_index", "1", "File Index") #not used
dbutils.widgets.text("total_files_len", "1", "total_files_len") #not used

dbutils.widgets.text("dbutils_thread_pool_size", "80", "DBUtils Thread Pool Size")

src_path = dbutils.widgets.get("src_path")
dest_path = dbutils.widgets.get("dest_path")
file_type = dbutils.widgets.get("file_type")
dbutils_thread_pool_size  = int(dbutils.widgets.get("dbutils_thread_pool_size"))
```

```python
print(f"Copying  {file_type} {src_path} to {dest_path} ")
```

```python
# The operation is parallelized by first recursively listing the objects in the source directory using a thread pool on the driver and subsequently launching a Spark job to perform the individual operations.

spark.conf.set("spark.databricks.service.dbutils.fs.parallel.enabled", True)
spark.conf.set("spark.databricks.service.dbutils.fs.parallel.ls.threadPoolSize", dbutils_thread_pool_size) #default 20
#spark.conf.set("spark.databricks.service.dbutils.fs.parallel.ls.timeoutSeconds", 7200)
```

```python
dbutils.fs.cp(src_path, dest_path, recurse=True if file_type == 'dir' else False)
```

```python
start_time = time.time()
result = False
try: 
    result = dbutils.fs.cp(src_path, dest_path, recurse=True if file_type == 'dir' else False)
except Exception as e:
    print(f"Exception {e}")
dbutils.notebook.exit(json.dumps( {"status": "SUCCESS" if result == True else "FAILED","duration": time.time() -start_time}))
```
