```python

from datetime import datetime
dbutils.widgets.text("catalog", "users")
dbutils.widgets.text("schema", "dom_rodrigues")
 
dbutils.widgets.text("s3_batch_copy_id", "Test1", "Batch Copy Id to compare")


s3_batch_copy_id=dbutils.widgets.get("s3_batch_copy_id")
catalog = dbutils.widgets.get("catalog")
schema =  dbutils.widgets.get("schema") 

debug =  True

insert_data_into_mapping_table = False
 
spark.sql(f"use {catalog}.{schema}")
```

```python

spark.sql(f"""select mapping.s3_batch_copy_id, mapping.runId, mapping.src_bucket_name, mapping.dest_bucket_name, mapping.last_recon_timestamp,
mapping.status,  recon.status, recon.time_stamp, count(*)
from {catalog}.{schema}.ucd_src_dest_mapping mapping
join {catalog}.{schema}.ucd_s3_inventory_recon recon 
on mapping.s3_batch_copy_id = recon.s3_batch_copy_id and mapping.runId = recon.runId
and mapping.enabled= True
group by mapping.s3_batch_copy_id, mapping.runId, mapping.src_bucket_name, mapping.dest_bucket_name, mapping.last_recon_timestamp,
mapping.status,  recon.status, recon.time_stamp
""").display() if debug else None
```

```python
spark.sql(f"""
select recon.*
from {catalog}.{schema}.ucd_src_dest_mapping mapping
join {catalog}.{schema}.ucd_s3_inventory_recon recon 
on mapping.s3_batch_copy_id = recon.s3_batch_copy_id and mapping.runId = recon.runId
and mapping.enabled= True
and mapping.s3_batch_copy_id = '{s3_batch_copy_id}'
and recon.status = 'EXISTS_IN_SRC_AND_BUT_NOT_IN_DEST'
""").display() if debug else None
```

```python
spark.sql(f"""
select mapping.src_bucket_name, mapping.s3_batch_copy_id, mapping.runId, recon.status, count(*)
from users.dom_rodrigues.ucd_src_dest_mapping mapping
join users.dom_rodrigues.ucd_s3_inventory_recon recon 
on mapping.s3_batch_copy_id = recon.s3_batch_copy_id and mapping.runId = recon.runId
and mapping.enabled= True
and mapping.s3_batch_copy_id = 'Test1'
group by mapping.src_bucket_name, mapping.s3_batch_copy_id, mapping.runId, recon.status
""").display() if debug else None
```
